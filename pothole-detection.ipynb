```python
with open('/kaggle/working/data.yaml', 'r') as f:
    file_path=yaml.safe_load(f)
print(file_path)
```

```python
from ultralytics import YOLO
model=YOLO('yolov8n.yaml').load('yolov8n.pt')
```

```python
## training the model #1-image_size, 2-data.yaml
images=glob.glob('/kaggle/input/pothole-detection-dataset/train/images/*')
img=cv2.imread(images[0])
imgsz=img.shape[0]
model.train(data='data.yaml',epochs=10,imgsz=imgsz,device=0)
```

```python
model.export()
```

```python
images=glob.glob('/kaggle/input/pothole-detection-dataset/valid/images/*')
results=model([images[1]],stream=False)
res=results[0].plot()
Image.fromarray(res)
```

```python
test_image_path=glob.glob("/kaggle/input/pothole-detection-dataset/valid/images/*.jpg")
test_model=YOLO("/kaggle/working/runs/detect/train/weights/best.pt")
results=test_model(test_image_path[:5],stream=False)
for i in range(4):
    res=results[i].plot()
    display(Image.fromarray(res))
```
